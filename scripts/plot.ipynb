```python
import matplotlib.pyplot as plt
import glob
import json
```

```python
log_pattern = "../output/BigEarthNet-v1_0_stack/*_ht_eps_010/010/beam_010_ops_6_split_5.log"
small_times, small_crs, medium_times, medium_crs = [], [], [], []
for f in glob.glob(log_pattern):
    with open(f, "r") as log_file:
        fields = log_file.read().split(",")
        time = float(fields[1])
        
        cr = float(fields[3])

        if "_187_" in f:
            medium_times.append(time)
            medium_crs.append(cr)
        elif "_18_" in f:
            small_times.append(time)
            small_crs.append(cr)
```

```python
log_pattern = "../output/BigEarthNet-v1_0_stack/*_ht_eps_010/010/beam_010_ops_6_split_0.log"
small_times_greedy, small_crs_greedy, medium_times_greedy, medium_crs_greedy = [], [], [], []
for f in glob.glob(log_pattern):
    with open(f, "r") as log_file:
        fields = log_file.read().split(",")
        time = float(fields[1])
        
        cr = float(fields[3])

        if "_187_" in f:
            medium_times_greedy.append(time)
            medium_crs_greedy.append(cr)
        elif "_18_" in f:
            small_times_greedy.append(time)
            small_crs_greedy.append(cr)
```

```python
log_pattern = "../output/BigEarthNet-v1_0_stack/*_test_[0-9]/010/beam_010_ops_6_split_5.log"
small_times_core, small_crs_core, medium_times_core, medium_crs_core = [], [], [], []
for f in glob.glob(log_pattern):
    with open(f, "r") as log_file:
        fields = log_file.read().split(",")
        time = float(fields[1])
        
        cr = float(fields[3])

        if "_187_" in f:
            medium_times_core.append(time)
            medium_crs_core.append(cr)
        elif "_18_" in f:
            small_times_core.append(time)
            small_crs_core.append(cr)
```

```python
log_pattern = "../output/BigEarthNet-v1_0_stack/*_test_[0-9]/010/beam_010_ops_6_split_0.log"
small_times_core_greedy, small_crs_core_greedy, medium_times_core_greedy, medium_crs_core_greedy = [], [], [], []
for f in glob.glob(log_pattern):
    with open(f, "r") as log_file:
        fields = log_file.read().split(",")
        time = float(fields[1])
        
        cr = float(fields[3])

        if "_187_" in f:
            medium_times_core_greedy.append(time)
            medium_crs_core_greedy.append(cr)
        elif "_18_" in f:
            small_times_core_greedy.append(time)
            small_crs_core_greedy.append(cr)
```

```python
plt.scatter(small_times, small_crs, label="from HT + error split", marker="+")
plt.scatter(small_times_greedy, small_crs_greedy, label="from HT + greedy error", marker="x")
plt.scatter(small_times_core, small_crs_core, label="from core + error split", marker="1")
plt.scatter(small_times_core_greedy, small_crs_core_greedy, label="from core + greedy error", marker="2")

plt.xlabel("Time (s)")
plt.ylabel("Compression")
plt.title("Shape: 18x120x120x12")
# plt.xscale("log")
plt.legend(loc="upper center")
plt.show()
```

```python
plt.scatter(medium_times, medium_crs, label="from HT + error split", marker="v")
plt.scatter(medium_times_greedy, medium_crs_greedy, label="from HT + greedy error", marker="s")
plt.scatter(medium_times_core, medium_crs_core, label="from core + error split", marker="o")
plt.scatter(medium_times_core_greedy, medium_crs_core_greedy, label="from core + greedy error", marker="*")

plt.xlabel("Time (s)")
plt.ylabel("Compression")
# plt.xscale("log")
plt.title("Shape: 17x11x120x120x12")
plt.legend()
plt.show()
```
